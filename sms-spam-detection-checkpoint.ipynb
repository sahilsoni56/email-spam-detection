```python
#data preprocessing
# .lower case
# .tokenization
# .remove special characters
# .remove stopwords and punctuations
# .stmming
```

```python
from nltk.corpus import stopwords
import string
from nltk.stem.porter import PorterStemmer
nltk.download('stopwords')

ps = PorterStemmer()

def transform_text(text):

    #lower the words
    text = text.lower()

    #tokenizing words
    text = nltk.word_tokenize(text)


#remove special charcters
    y = []
    for i in text:
        if i.isalnum:
            y.append(i)


    text = y[:]
    y.clear()


    #reoving stopwords and punctations
    for i in text:
       if i not in stopwords.words('english') and i not in string.punctuation:
           y.append(i)


    #stemming
    text = y[:]
    y.clear()
    for i in text:
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
df['transformed_text'] = df['text'].apply(transform_text)
df.head(10)

```

```python
#MODEL BUILDING

#NAIVE BAYES FOR TEXUAL DATA(best algo)

from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
# cv= CountVectorizer()
tf = TfidfVectorizer(max_features=3000)
```

```python
# x = cv.fit_transform(df['transformed_text']).toarray()
x = tf.fit_transform(df['transformed_text']).toarray()
y = df['target'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
#models

from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,precision_score
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python

gnb.fit(x_train,y_train)
ypredict1 = gnb.predict(x_test)
print("Accuracy of the model is : ",accuracy_score(y_test , ypredict1))
print("precision of this model is :",precision_score(y_test , ypredict1))


#from this we know that gausainNb is not good for this dataset
#lets try another model to train
```

```python
mnb.fit(x_train,y_train)
ypredict2 = mnb.predict(x_test)
print("Accuracy of the model is : ",accuracy_score(y_test , ypredict2))
print("precision of this model is :",precision_score(y_test , ypredict2))


#lets try another one also
```

```python
bnb.fit(x_train,y_train)
ypredict3 = bnb.predict(x_test)
print("Accuracy of the model is : ",accuracy_score(y_test , ypredict3))
print("precision of this model is :",precision_score(y_test , ypredict3))
```
